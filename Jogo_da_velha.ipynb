```python
# Marca no quadro

def marcando_posicao(linha1,linha2,linha3,posicao,jogador):
    if posicao==1:
        linha1[0] = jogador
    if posicao==2:
        linha1[1] = jogador
    if posicao==3:
        linha1[2] = jogador
    if posicao==4:
        linha2[0] = jogador
    if posicao==5:
        linha2[1] = jogador
    if posicao==6:
        linha2[2] = jogador
    if posicao==7:
        linha3[0] = jogador
    if posicao==8:
        linha3[1] = jogador
    if posicao==9:
        linha3[2] = jogador
```
